```r
#Medidas de resumen:
summary(data) # calculamos todas las medias de dispersion y de resumen y localización.
# calcular medidas de dispersión:

varCRIM= var(data$crim)
varZN = var(data$zn)
varINDUS = var(data$indus)
varNOX = var(data$nox)
varRM = var(data$rm)
varAGE = var(data$age)
varDIS = var(data$dis)
varRAD = var(data$rad)
varTAX = var(data$tax)
varPTRATIO = var(data$ptratio)
varB = var(data$b)
varLSTAT = var(data$lstat)
varMEDV = var(data$medv)

# mostrar las varianzas, desviaciones estandar y coeficientes de variacion
cat("\n")
cat("varianza CRIM: ",varCRIM, "\n")
cat("desviacion estandar CRIM: ", sqrt(varCRIM), "\n")
cat("coeficiente de variación CRIM: ", sqrt(varCRIM)/mean(data$crim), "\n")
cat("\n")
cat("varianza ZN: ", varZN, "\n")
cat("desviacion estandar ZN: ", sqrt(varZN), "\n")
cat("coeficiente de variación ZN: ", sqrt(varZN)/mean(data$zn), "\n")
cat("\n")
cat("varianza INDUS: ", varINDUS, "\n")
cat("desviacion estandar INDUS: ", sqrt(varINDUS), "\n")
cat("coeficiente de variación INDUS: ", sqrt(varINDUS)/mean(data$indus), "\n")
cat("\n")
cat("varianza NOX: ", varNOX, "\n")
cat("desviacion estandar NOX: ", sqrt(varNOX), "\n")
cat("coeficiente de variación NOX: ", sqrt(varNOX)/mean(data$nox), "\n")
cat("\n")
cat("varianza RM: ", varRM, "\n")
cat("desviacion estandar RM: ", sqrt(varRM), "\n")
cat("coeficiente de variación RM: ", sqrt(varRM)/mean(data$rm), "\n")
cat("\n")
cat("varianza AGE: ", varAGE, "\n")
cat("desviacion estandar AGE: ", sqrt(varAGE), "\n")
cat("coeficiente de variación AGE: ", sqrt(varAGE)/mean(data$age), "\n")
cat("\n")
cat("varianza DIS: ", varDIS, "\n")
cat("desviacion estandar DIS: ", sqrt(varDIS), "\n")
cat("coeficiente de variación DIS: ", sqrt(varDIS)/mean(data$dis), "\n")
cat("\n")
cat("varianza RAD: ", varRAD, "\n")
cat("desviacion estandar RAD: ", sqrt(varRAD), "\n")
cat("coeficiente de variación RAD: ", sqrt(varRAD)/mean(data$rad), "\n")
cat("\n")
cat("varianza TAX: ", varTAX, "\n")
cat("desviacion estandar TAX: ", sqrt(varTAX), "\n")
cat("coeficiente de variación TAX: ", sqrt(varTAX)/mean(data$tax), "\n")
cat("\n")
cat("varianza PTRATIO: ", varPTRATIO, "\n")
cat("desviacion estandar PTRATIO: ", sqrt(varPTRATIO), "\n")
cat("coeficiente de variación PTRATIO: ", sqrt(varPTRATIO)/mean(data$ptratio), "\n")
cat("\n")
cat("varianza B: ", varB, "\n")
cat("desviacion estandar B: ", sqrt(varB), "\n")
cat("coeficiente de variación B: ", sqrt(varB)/mean(data$b), "\n")
cat("\n")
cat("varianza LSTAT: ", varLSTAT, "\n")
cat("desviacion estandar LSTAT: ", sqrt(varLSTAT), "\n")
cat("coeficiente de variación LSTAT: ", sqrt(varLSTAT)/mean(data$lstat), "\n")
cat("\n")
cat("varianza MEDV: ", varMEDV, "\n")
cat("desviacion estandar MEDV: ", sqrt(varMEDV), "\n")
cat("coeficiente de variación MEDV: ", sqrt(varMEDV)/mean(data$medv), "\n")
```

```r
# cálculo de curtosis y asimetría
library(moments)
cat("NOX", "\n")
cat("Asimetría: ", skewness(data$nox, na.rm = TRUE), "\n")
cat("Curtosis: ", kurtosis(data$nox, na.rm = TRUE)+3, "\n")
cat("INDUS", "\n")
cat("Asimetría: ", skewness(data$indus, na.rm = TRUE), "\n")
cat("Curtosis: ", kurtosis(data$indus, na.rm = TRUE)+3, "\n")
cat("LSTAT", "\n")
cat("Asimetría: ", skewness(data$lstat, na.rm = TRUE), "\n")
cat("Curtosis: ", kurtosis(data$lstat, na.rm = TRUE)+3, "\n")
```

**Comentarios**

- **NOX:** La asimetría dió un valor mayor que cero, por ende la cola está hacia la derecha. La curtosis en este caso es leptocurtica ya que la gráfica tiene una punta bastante alta y su curtosis es mayor que 3 por esto se concluye que es del tipo leptocúrtica. Con esto podemos decir que si concuerda con lo que presenta el gráfico.

- **INDUS:** La asimetría dio un valor mayor que cero, por lo tanto, la cola está levemente hacia la derecha. La curtosis es también mayor que 3, por lo que la distribución se considera leptocúrtica, aunque de forma moderada. Esto significa que hay una mayor concentración de valores cercanos a la media, pero también algunos valores extremos. Se puede decir que la descripción concuerda con la forma del gráfico, que muestra una ligera inclinación hacia valores más altos y una forma picuda.

- **LSTAT:** La asimetría dio un valor claramente mayor que cero, lo que indica una cola prolongada hacia la derecha. La curtosis es considerablemente mayor que 3, lo cual indica que la distribución es fuertemente leptocúrtica, es decir, con una alta concentración de datos cerca de la media y colas más pesadas. Esto concuerda con el gráfico de densidad, donde se observa un pico pronunciado y una distribución sesgada hacia valores altos.

**Comentarios**

- Se escoge lstat y crim, porque puede que la tasa de criminalidad y el porcentaje de poblacion de menor status puede variar según la ubicación geográfica (cerca del río).
